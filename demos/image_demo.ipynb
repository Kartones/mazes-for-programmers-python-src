```python
# Temporarilly add parent folder to path
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from base.colored_grid import ColoredGrid
from algorithms.binary_tree import BinaryTree
from exporters.pixel_exporter import PixelExporter
from exporters.png_exporter import PNGExporter
```

```python
# Run BinaryTree
grid = ColoredGrid(100, 100)
grid = BinaryTree.on(grid)
```

```python
# Color cells
start_cell = grid[grid.rows//2, grid.cols//2]
grid.distances = start_cell.distances
```

```python
# Use PNG exporter...
pe = PNGExporter()
im1 = pe._render(grid, coloring=True)
im1
```

```python
# ... or the pixel exporter
pe = PixelExporter()
im2 = pe._render(grid, coloring=True)
im2
```

```python
# Run this cell to save the images
im1.save('PNGExporter_maze.png')
im2.save('PIxelExporter_maze.png')
```
